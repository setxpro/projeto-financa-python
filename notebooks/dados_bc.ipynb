```python
from datetime import datetime, date
from matplotlib import pyplot as pl
import numpy as np
from bcb import sgs
```

## Passo 2 - Coletar dados do usuário 

```python
capital = float(input("Digite o capital investido: ")) 
frequencia = input("Digite a frequência do período (Y, M, D): ")
inicio = input("Digite a data inicial maior do que 1995/01/01 no formato YYYY/MM/DD: ")
final = input("Digite a data final no seguinte formato YYYY/MM/DD: ")
```

## Passo 3 - Tratar dados coletados

```python
data_inicio = datetime.strptime(inicio, "%Y/%m/%d").date()
data_final = datetime.strptime(final, "%Y/%m/%d").date()
```

## Passo 4 - Pegar dados da SELIC do Banco Central.

```python
taxa_selic = sgs.get({"selic": 11}, start=data_inicio, end=data_final)

taxa_selic = taxa_selic/100

taxa_selic
```

## Passo 5 - Calcular retorno do período

```python
capital_acumulado = capital * (1 + taxa_selic["selic"]).cumprod() - 1

capital_acumulado
```

```python
# Frequencia Diária para mensal
capital_com_frequencia = capital_acumulado.resample(frequencia).last()

capital_com_frequencia
```

### Filtrar dados da SELIC no período

```python
data_inicial = date(2000, 1, 1)
data_final = date(2022, 3, 31)
```

```python
selic_question = sgs.get({"selic": 11}, start=data_inicial, end=data_final)/100

selic_question
```

### Calcular rentabilidade das janelas de 500 dias.

```python
# Aplica na janela de 500 dias a rentabilidade acumulada.
janelas_500_dias = ((1 + selic_question).rolling(window = 500).apply(np.prod) + 1)

janelas_500_dias                
```

```python
janelas_500_dias = janelas_500_dias.reset_index()

janelas_500_dias
```

## Criar range de datas na tabela

```python
janelas_500_dias["data_inicial"] = janelas_500_dias["Date"].shift(500)

janelas_500_dias
```

```python
# Remove os dados faltantes
janelas_500_dias = janelas_500_dias.dropna()

janelas_500_dias
```

```python
janelas_500_dias.columns = ["data_final", "retorno_selic_500d", "data_inicial"]

janelas_500_dias
```

```python
# Pegar o maior retorno da tabela
maior_retorno = janelas_500_dias["retorno_selic_500d"].max()

maior_retorno
```

```python
# Linha com o maior retorno
row = janelas_500_dias[janelas_500_dias["retorno_selic_500d"] == maior_retorno]

row
```
